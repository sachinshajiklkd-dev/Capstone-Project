# Step 27

```python
for col in cat_cols:
    plt.figure(figsize=(6,6))
    df[col].value_counts().plot(
        kind='pie',
        autopct='%1.1f%%',
        startangle=90
    )
    plt.ylabel("")
    plt.title(f'Pie Chart of {col}')
    plt.show()
```

# Step 28

```python
if len(num_cols) >= 2:
    for i in range(len(num_cols)-1):
        plt.figure(figsize=(6,4))
        sns.scatterplot(
            data=df,
            x=num_cols[i],
            y=num_cols[i+1]
        )
        plt.title(f'{num_cols[i]} vs {num_cols[i+1]}')
        plt.show()
```

# Step 29

```python
if len(num_cols) > 1:
    plt.figure(figsize=(10,8))
    sns.heatmap(
        df[num_cols].corr(),
        annot=True,
        cmap='coolwarm',
        fmt='.2f'
    )
    plt.title("Correlation Heatmap")
    plt.show()
```

# Step 30

```python
if len(num_cols) >= 2:
    plt.figure(figsize=(8,5))
    plt.plot(df[num_cols[0]], label=num_cols[0])
    plt.title(f'Line Plot of {num_cols[0]}')
    plt.xlabel("Index")
    plt.ylabel(num_cols[0])
    plt.legend()
    plt.show()
```

# Step 31

```python
if len(num_cols) >= 4:
    fig, axes = plt.subplots(2,2, figsize=(12,8))

    for ax, col in zip(axes.flatten(), num_cols[:4]):
        sns.histplot(df[col], kde=True, ax=ax)
        ax.set_title(col)

    plt.tight_layout()
    plt.show()
```
